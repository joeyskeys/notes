## Specular lobe

Two specular lobes are included:
 1. Main microfacet specular lobe, in 2012 BRDF it's a specular BRDF, in 2015 BSDF it combinded with BTDF, check [Bruce Walter 2007].
 2. Secondary specular as clearcoat, only contributing to reflection.
 
And in the extended BSDF, surface absorption and volumetric absorption are added:
 1. surface absorption : "we multiply the refraction result by the square root of the surface color", how to get the square root of a color?
 2. "we apply volumetric absorption based on the length of refracted paths through the solid", according to Beer-Lambert law, $T = e^{-\sigma_\alpha d}$, transmittance color is used as parameter, we calculate $\sigma_\alpha = -(\log{T})/d$
 
## Diffuse lobe

In 2012 BRDF, diffuse lobe adds a Fresnel factor : $(1 - F(\theta_l))(1 - F(\theta_d))$, where $F(\theta)$ is the fresnel factor for reflection.
And diffuse lobe is modeled like :
$$
f_d = \frac{baseColor}{\pi} (1 + (F_{D90} - 1)(1 - \cos{\theta_l})^5)(1 + (F_{D90} - 1)(1 - \cos{\theta_v})^5)
$$
where
$$
F_{D90} = 0.5 + 2 * roughenss * \cos^2{\theta_d}
$$

In 2015 BSDF, diffuse lobe is refactored into directional microfacet effects and non-directional subsurface effects, then the non-directional part can either be a diffusion model or a volumetric scattering model.
$$
f_d = f_{lambert}(1 - 0.5F_L)(1 - 0.5F_V) + f_{retro-reflection}
$$
$$
f_{lambert} = \frac{baseColor}{\pi}
$$
$$
f_{retro_reflection} = \frac{baseColor}{\pi} R_R (F_L + F_V + F_L F_V(R_R - 1))
$$
where
$$
F_L = (1 - \cos{\theta_l})^5
$$
$$
F_V = (1 - \cos{\theta_v})^5
$$
$$
R_R = 2 roughness \cos^2{\theta_d}
$$

## Sheen lobe

Sheen is a additive part of diffuse lobe. Tinted towards baseColor by sheenTint, has the form of $sheen * (1 - \cos{\theta_d})^5$.
According the implementations, sheen part is a seperate diffuse lobe. To be investigated.
